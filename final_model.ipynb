```python
# ===============================
# Complete SARIMAX Forecast Code
# With Auto_ARIMA, Result Logging, and Parallelization
# ===============================

# --- Imports ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.statespace.sarimax import SARIMAX
from pmdarima import auto_arima
from joblib import Parallel, delayed
import warnings
import json
warnings.filterwarnings("ignore")

# --- Load Data ---
file_path = "Merged_data_final.csv"
df = pd.read_csv(file_path)
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df.dropna(subset=['Date'], inplace=True)
df.sort_values('Date', inplace=True)

# --- Feature Engineering ---
def add_engineered_features(stock_df):
    stock_df = stock_df.copy()
    stock_df['EMA_12'] = stock_df['Close'].ewm(span=12, adjust=False).mean()
    stock_df['EMA_26'] = stock_df['Close'].ewm(span=26, adjust=False).mean()
    stock_df['MACD'] = stock_df['EMA_12'] - stock_df['EMA_26']
    stock_df['MACD_Signal'] = stock_df['MACD'].ewm(span=9, adjust=False).mean()
    delta = stock_df['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
    rs = gain / loss
    stock_df['RSI_14'] = 100 - (100 / (1 + rs))
    stock_df['Lag_Close'] = stock_df['Close'].shift(1)
    stock_df['Price_Change_Pct'] = stock_df['Close'].pct_change()
    stock_df['Volatility_7d'] = stock_df['Close'].rolling(window=7).std()
    stock_df['Weekday'] = stock_df['Date'].dt.weekday
    return stock_df

# --- Auto ARIMA Order Suggestion ---
def suggest_sarimax_order(endog, exog):
    try:
        model = auto_arima(endog, exogenous=exog, seasonal=False, stepwise=True, 
                           suppress_warnings=True, max_order=6, maxiter=30)
        return model.order
    except Exception as e:
        print(f"[Auto-ARIMA Error] {e}")
        return None

# --- Per-stock Modeling Function ---
def process_stock(stock):
    print(f"\n===== Processing {stock} =====")
    stock_df = df[df['Stock'] == stock].copy()
    stock_df = add_engineered_features(stock_df)
    stock_df.set_index('Date', inplace=True)
    stock_df = stock_df.sort_index()

    feature_cols = ['sentiment_mean', 'Volume', 'sentiment_count',
                    'Price_Change_Pct', 'Lag_Close', 'Volatility_7d', 'Weekday',
                    'EMA_12', 'EMA_26', 'MACD', 'MACD_Signal', 'RSI_14']

    ts_df = stock_df[['Close'] + feature_cols].dropna()

    if len(ts_df) < 50:
        print(f"Not enough data for {stock}. Skipping.")
        return None

    endog = ts_df['Close']
    exog = ts_df[feature_cols]
    exog_scaled = pd.DataFrame(StandardScaler().fit_transform(exog),
                               index=exog.index, columns=exog.columns)

    # Try auto_arima suggestion
    order_grid = [(2, 1, 2), (3, 1, 3)]  # default
    suggested_order = suggest_sarimax_order(endog, exog_scaled)
    if suggested_order and suggested_order not in order_grid:
        print(f"üîç Suggested Order from auto_arima: {suggested_order}")
        order_grid = [suggested_order] + order_grid

    best_rmse = float('inf')
    best_forecast = None
    best_order = None
    best_aic = None
    best_bic = None
    accuracy = None
    true_values = None

    for order in order_grid:
        try:
            model = SARIMAX(endog, exog=exog_scaled, order=order, enforce_stationarity=False)
            result = model.fit(disp=False, maxiter=50)
            forecast = result.forecast(steps=10, exog=exog_scaled.iloc[-10:])
            true_values = endog[-10:]
            rmse = np.sqrt(mean_squared_error(true_values, forecast))

            train_forecast = result.fittedvalues[-len(true_values):]
            train_rmse = np.sqrt(mean_squared_error(true_values, train_forecast))
            rmse_gap = rmse - train_rmse

            aic = result.aic
            bic = result.bic

            print(f"Order {order} - RMSE: {rmse:.4f}, Train RMSE: {train_rmse:.4f}, Gap: {rmse_gap:.4f}, AIC: {aic:.2f}, BIC: {bic:.2f}")
            if rmse_gap > 3:
                print("‚ö†Ô∏è  Potential overfitting detected (high RMSE gap)")

            if rmse < best_rmse:
                best_rmse = rmse
                best_forecast = forecast
                best_order = order
                best_aic = aic
                best_bic = bic
        except Exception as e:
            print(f"Error with order {order} for {stock}: {e}")
            continue

    if best_forecast is not None:
        valid_idx = true_values != 0
        if valid_idx.sum() > 0:
            true_vals = true_values[valid_idx].to_numpy()
            forecast_vals = best_forecast[-len(true_vals):].to_numpy()
            mape = np.mean(np.abs((true_vals - forecast_vals) / true_vals)) * 100
            accuracy = 100 - mape
            print(f"üìä Forecast Accuracy for {stock}: {accuracy:.2f}%")

        print(f"\n‚úÖ Best model for {stock}: Order {best_order}, RMSE: {best_rmse:.4f}, AIC: {best_aic:.2f}, BIC: {best_bic:.2f}")

        # Plot actual vs forecast
        plt.figure(figsize=(10, 5))
        plt.plot(endog[-20:], label='Actual')
        forecast_index = endog.index[-10:]
        plt.plot(forecast_index, best_forecast, label='Forecast', linestyle='--')
        plt.title(f'{stock} - SARIMAX Forecast (Order {best_order})')
        plt.xlabel('Date')
        plt.ylabel('Close Price')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

        return {
            "Stock": stock,
            "Best_Order": best_order,
            "Best_RMSE": round(best_rmse, 4),
            "Best_AIC": round(best_aic, 2),
            "Best_BIC": round(best_bic, 2),
            "Forecast_Accuracy": round(accuracy, 2) if accuracy is not None else None
        }
    else:
        print(f"‚ùå No valid model found for {stock}")
        return None

# --- Run for All Stocks (Sequential or Parallel) ---
stocks = df['Stock'].unique()

# üîÅ Option A: Sequential
results = [res for res in map(process_stock, stocks) if res]

# üîÅ Option B: Parallel (uncomment below if needed)
# results = Parallel(n_jobs=-1)(delayed(process_stock)(stock) for stock in stocks)
# results = [res for res in results if res]

# --- Save Results ---
results_df = pd.DataFrame(results)
results_df.to_csv("sarimax_stock_results.csv", index=False)
with open("sarimax_stock_results.json", "w") as f:
    json.dump(results, f, indent=4)

print("\nüìÅ Results saved to 'sarimax_stock_results.csv' and 'sarimax_stock_results.json'")
```

```python
# -----------------------
# Imports
# -----------------------
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from statsmodels.tsa.statespace.sarimax import SARIMAX
from pmdarima import auto_arima
import warnings
import json

warnings.filterwarnings("ignore")

# -----------------------
# Load 5-minute Interval Data
# -----------------------
file_path = "with_faded_sentiment_60.csv"  # Update this path as needed
df = pd.read_csv(file_path)

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df.dropna(subset=['Date'], inplace=True)
df.sort_values('Date', inplace=True)

# -----------------------
# Forecast Horizons (excluding 5min)
# -----------------------
forecast_steps = {
    "15min": 3,
    "30min": 6,
    "1hour": 12
}

# -----------------------
# Feature Engineering
# -----------------------
def add_intraday_features(df):
    df = df.copy()
    df['EMA_5'] = df['Close'].ewm(span=5, adjust=False).mean()
    df['EMA_12'] = df['Close'].ewm(span=12, adjust=False).mean()
    df['Volatility_30min'] = df['Close'].rolling(window=6).std()
    df['Price_Change_Pct'] = df['Close'].pct_change()
    df['Lag_Close'] = df['Close'].shift(1)
    df['Weekday'] = df['Date'].dt.weekday
    return df

# -----------------------
# Forecasting Function
# -----------------------
def process_stock_intraday(stock):
    print(f"\n===== Processing {stock} =====")
    stock_df = df[df['Stock'] == stock].copy()
    stock_df = add_intraday_features(stock_df)
    stock_df.set_index('Date', inplace=True)
    stock_df = stock_df.sort_index()

    feature_cols = [
        'Volume',
        'faded_sentiment_score',
        'EMA_5',
        'EMA_12',
        'Volatility_30min',
        'Price_Change_Pct',
        'Lag_Close',
        'Weekday'
    ]

    ts_df = stock_df[['Close'] + feature_cols].dropna()

    if len(ts_df) < 100:
        print(f"Not enough intraday data for {stock}. Skipping.")
        return []

    # Trim last 12 rows to ensure future price is available for all forecasts
    ts_df = ts_df.iloc[:-12]

    endog = ts_df['Close']
    exog = ts_df[feature_cols]
    exog_scaled = pd.DataFrame(StandardScaler().fit_transform(exog),
                               index=exog.index, columns=exog.columns)

    # Auto ARIMA order suggestion
    try:
        auto_model = auto_arima(endog, exogenous=exog_scaled, seasonal=False,
                                stepwise=True, suppress_warnings=True, max_order=6, maxiter=30)
        order = auto_model.order
        print(f"üîç Suggested Order from auto_arima: {order}")
    except:
        order = (2, 1, 2)
        print(f"‚ö†Ô∏è Auto_ARIMA failed. Using fallback order {order}")

    try:
        model = SARIMAX(endog, exog=exog_scaled, order=order, enforce_stationarity=False)
        result = model.fit(disp=False, maxiter=50)
    except Exception as e:
        print(f"‚ùå Model failed for {stock}: {e}")
        return []

    current_price = endog.iloc[-1]
    now_time = endog.index[-1]
    results = []

    # --- Overfitting Check ---
    try:
        train_pred = result.fittedvalues[-12:]
        train_rmse = np.sqrt(np.mean((train_pred - endog[-12:]) ** 2))
    except:
        train_rmse = np.nan

    for label, steps in forecast_steps.items():
        try:
            if len(endog) < steps + 1:
                print(f"‚ö†Ô∏è Not enough data to forecast {label} for {stock}")
                continue

            forecast = result.forecast(steps=steps, exog=exog_scaled.iloc[-steps:])
            forecast_mean = forecast.mean()
            predicted_direction = "Up" if forecast_mean > current_price else "Down"

            actual_future_price = endog.iloc[-1 + steps]
            actual_direction = "Up" if actual_future_price > current_price else "Down"
            trend_accuracy = 1 if predicted_direction == actual_direction else 0

            # MAPE, RMSE, Confidence
            mape = np.mean(np.abs((forecast - current_price) / current_price)) * 100
            rmse = np.sqrt(np.mean((forecast - current_price)**2))
            rmse_gap = rmse - train_rmse
            confidence = round(100 - mape, 2)

            print(f"üìà {label} | Predicted: {forecast_mean:.2f}, Actual: {actual_future_price:.2f}, "
                  f"Trend: {predicted_direction} vs {actual_direction}, "
                  f"Accuracy: {trend_accuracy}, Confidence: {confidence}%, "
                  f"RMSE: {rmse:.4f}, Train RMSE: {train_rmse:.4f}, Gap: {rmse_gap:.4f}")

            if rmse_gap > 3:
                print("‚ö†Ô∏è Overfitting Warning: RMSE gap is too high.")

            results.append({
                "Stock": stock,
                "Forecast_Horizon": label,
                "Timestamp": now_time.strftime("%Y-%m-%d %H:%M"),
                "Current_Price": round(current_price, 4),
                "Forecast_Avg": round(forecast_mean, 4),
                "Predicted_Trend": predicted_direction,
                "Actual_Trend": actual_direction,
                "Trend_Accuracy": trend_accuracy,
                "Confidence (%)": confidence,
                "MAPE": round(mape, 2),
                "RMSE": round(rmse, 4),
                "Train_RMSE": round(train_rmse, 4),
                "RMSE_Gap": round(rmse_gap, 4),
                "Model_Order": order
            })
        except Exception as e:
            print(f"‚ö†Ô∏è Forecast failed for {label}: {e}")

    return results

# -----------------------
# Run Forecasting
# -----------------------
stocks = df['Stock'].unique()
all_results = []

for stock in stocks:
    result = process_stock_intraday(stock)
    all_results.extend(result)

# -----------------------
# Save Output
# -----------------------
results_df = pd.DataFrame(all_results)
results_df.to_csv("robust_sarimax_forecasts_no_5min.csv", index=False)

with open("robust_sarimax_forecasts_no_5min.json", "w") as f:
    json.dump(all_results, f, indent=4)

print("\n‚úÖ Forecasts (15, 30, 60 min) saved with confidence and overfitting validation.")
```
